# even with for loop

```python
b =[1,2,3,4,5,6,7,8,10]
size = len(b)
size
for i in range(size):
    if b[i] % 2==0:
        print(f'b[{i}] : {b[i]}')
        
```

# max min

```python
f = [1,2,3,4,5,6,7,8,9]
min = f[0]
max = f[0]
for i in f:
    if i <min:
        min = i
    if i>max :
        max = i
print("min is" ,min)
print("max is" ,max)
```
